```python
# Import necessary modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from plotnine import *
```

```python
# Read in the data
df = pd.read_csv('Autism-Adult-Data.csv')
```

```python
# Make wide enough to see all columns and examine head of the data
pd.set_option('display.max_columns', df.shape[0]+1)

df.head()
```

```python
# Examine number of columns/rows
df.shape
```

```python
# Examine what patient information we have obtained from questionnaire
df.columns
```

```python
# Examine measures of central tendency
df.describe()
```

```python
# Examine dataframe for appropriate datatypes
df.dtypes
```

```python
# Change dtype for our Age for better usage in ML. 

# When attempting to change age will find '?' inside of our Age column.
# df.Age.value_counts()

# Change to numerical and force '?'' to Nan
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Replace NaN with mean
df.loc[df['Age'].isna(),'Age']  = df['Age'].mean()
```

```python
# With numerical data examine the spread of the numbers. 
# df.Age.describe()
# Note that the max is too large 383

# Replace max age with mean
# Re-examine spread of data
# Note to self: Fix hardcoding here

df.loc[df['Age'].eq(383),'Age']  = df['Age'].mean()
df.Age.describe()
```

```python
df.Ethnicity.value_counts()
```

For ease of usage I would like to edit some of the types of answers within some of the language based columns

```python
# For our purposes there is no discerneable difference for us between '?', 'Others', and 'Other' 
# Reclassify as just one type

df.Ethnicity = df.Ethnicity.replace({'?':'Others','others':'Others'})
```

```python
# Two ethnicities have extra quotation marks

for x in range(len(df)):
    df['Ethnicity'][x] = df['Ethnicity'][x].replace("'",'') 
```

```python
#Check for any issues 
df.Ethnicity.unique()
```

```python
# Double check there aren't any misspelling issues or multiple classifications for one gender 
# i.e. 'f' vs 'F' vs 'Fem'
df.Gender.unique()
```

```python
# From header I can see United States is oddly formated. 
#Checking all of the different types.
df.Country_of_res.unique()
```

```python
# Several countries have extra quotation marks
for x in range(len(df)):
    df['Country_of_res'][x] = df['Country_of_res'][x].replace("'",'') 
```

```python
# Check removal of quotations 
df.Country_of_res.unique()
```

```python
# Check for missing data files
df.isnull().sum()
```

```python
# A convoluted way to make sure that App_result == The summation of a 1 on each question 

scoreList = ['A1_Score','A2_Score', 'A3_Score', 'A4_Score', 'A5_Score', 'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'A10_Score']
scoreDF = df[scoreList]
scoreDF['Sum'] = df.loc[:, scoreList].sum(axis=1)
scoreDF.head()

scoreDF['equal'] = df.App_result.eq(scoreDF.Sum)
scoreDF.equal.value_counts()
```

```python
# Confirmed that App_result is a summation of how many questions the patient got assigned a '1' 
# So that is why the scores are between 0-10 as we saw when we ran .describe()
df.App_result.unique()
```

```python
# On questionnaire they ask if the patient is born with jaundice
df = df.rename(columns={"Jundice": "Jaundice"})
```

```python
# Change yes/no categorical varibles to binary 0/1
# Since it's simple with only 2 answer responses this is neater than get_dummies

df.ASD = df.ASD.map({'YES': 1, 'NO': 0})
df.Jaundice = df.Jaundice.map({'yes': 1, 'no': 0})
df.Family = df.Family.map({'yes': 1, 'no': 0})
df.Used_app_before = df.Used_app_before.map({'yes': 1, 'no': 0})

# Create new column for an encoded gender with 1 being male and 0 being female for ease of data exploration
df['GenderBinary'] = df.Gender.map({'m': 1, 'f': 0})
```

## 2. Exploratory Data Analysis

### EDA: Gender Differences
Examine the difference in diagnoses between the genders

```python
# Lets compare how the diagnosis rate varies between genders

sns.countplot(x='Gender', hue = 'ASD', data = df)
```

```python
#Find the actual count of each Gender that filled survey
femaleTotal = df[df['Gender']=='f']['Gender'].count()
maleTotal = df[df['Gender']=='m']['Gender'].count()

#Find the actual count of each Gender that filled survey and positive for ASD
femalePos = df[(df['Gender']=='f') & (df['ASD'] == 1)]['Gender'].count()
malePos = df[(df['Gender']=='f') & (df['ASD'] == 1)]['Gender'].count()

#Find percentage
femalePer = femalePos/femaleTotal
malePer = malePos/maleTotal
```

```python
#Coding a print out to give actual numbers for above countplot

print('There were ', femaleTotal, ' females represented in the survey. With ', '{0:.0%}'.format(femalePer), ' of women being diagnosed with Autism Spectrum Disorder.')
print('There were ', maleTotal, ' males represented in the survey. With ', '{0:.0%}'.format(malePer), ' of men being diagnosed with Autism Spectrum Disorder.')
```

It would appear that there is a slightly higher diagnosis for women with ASD than women. Can it be attributed to clinical biaseses, due to women having higher incidinces of austism, or should the 3% be considered negligible.

# Machine Learning

## 3. Preprocessing and Feature Selection

```python
# import machine learning packages

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df.columns
```

```python
df.Ethnicity.nunique()
```

```python
# Preprocessing. One hot encode (getdummies)the variables 'Ethnicity' and 'Relation'
#df = pd.get_dummies(df,columns=['Ethnicity', 'Relation'],drop_first=True)

# After running the below model with 'Ethnicity and Relation' they had a nearly negligible affect 
# on the predictive effect of the data. So we will not use them.
# There are too many countries with too small of a representation to be a good predictor of ASD diagnosis

# After doing a runthrough and reading the study. 
# It appears the 10 questions are very helpful in screening for ASD diagnosis. 
# They are so helpful the other 10 self-reported factors become less relevant.
# Not so many features PCA is necessary to determine feature reduction
```

```python
# Drop Age_desc since it has only one category and doesn't add value to our machine learning
df = df.drop('Age_desc', axis = 1)
df.head()
```

```python
# Create a list of features
# Dropping unneeded categorical columns 

features_list = list(df.drop(['Gender','ASD', 'Country_of_res','Ethnicity', 'Relation'], axis = 1).columns)
```

```python
# Assign X & y values
X = df[features_list]
y = df['ASD']
```

```python
#Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Machine Learning
## 4. Random Forest Model

```python
# Import for scoring features
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,roc_auc_score
from sklearn.metrics import accuracy_score,log_loss
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing

# Import for models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# Import for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
# Build random forest model
rf = RandomForestClassifier(random_state=42)

# Train the model and predict
model = rf.fit(X_train, y_train)
y_pred = model.predict(X_test)

y_pred_prob = model.predict_proba(X_test)
lr_probs = y_pred_prob[:,1]


#Find accuracy
ac = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
cm = confusion_matrix(y_test, y_pred)

print('Random Forest: Accuracy=%.3f' % (ac))
print('Random Forest: f1-score=%.3f' % (f1))
```

```python
print(classification_report(y_test, rf.predict(X_test)))
```

```python
# Display confusion matrix
print(cm)
```

```python
#print(roc_auc_score(y_test, y_pred))
```

```python
# Specific Numbers for feature importance

fi = pd.DataFrame({'feature': list(X_train.columns),
                   'importance': model.feature_importances_}).\
                    sort_values('importance', ascending = False)

# Display
fi.head()
```

```python
# Graph of feature importance
feature_importance = rf.feature_importances_

# Make importances relative to max importance

feature_importance = 100.0 * (feature_importance / feature_importance.max())[:30]
sorted_idx = np.argsort(feature_importance)[:30]

pos = np.arange(sorted_idx.shape[0]) + .5
print(pos.size)
sorted_idx.size
plt.figure(figsize=(10,10))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, X.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

```python
# 100% accuracy?
# Try again with cross validation
# Evaluate a logistic regression model using k-fold cross-validation
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
# prepare the cross-validation procedure
cv = KFold(n_splits=10, random_state=1, shuffle=True)
# create model
model = RandomForestClassifier()
# evaluate model
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

## KNN Model

```python
param_grid = {'n_neighbors':np.arange(1,50)}
knn = KNeighborsClassifier()
knn_cv= GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(X,y)
```

```python
print("Best Score:" + str(knn_cv.best_score_))
print("Best Parameters: " + str(knn_cv.best_params_))
```

```python
#Build KNN model with best params
knn = KNeighborsClassifier(n_neighbors = 13)

#Train the model
knnmodel = knn.fit(X_train, y_train)
y_pred_knn = knnmodel.predict(X_test)
y_pred_prob_knn = knnmodel.predict_proba(X_test)
lr_probs_knn = y_pred_prob_knn[:,1]
ac_knn = accuracy_score(y_test, y_pred_knn)


#Find accuracy
f1_knn = f1_score(y_test, y_pred_knn, average='weighted')
cm_knn = confusion_matrix(y_test, y_pred_knn)

print('K Nearest Neighbors: Accuracy=%.3f' % (ac_knn))

print('K Nearest Neighbors: f1-score=%.3f' % (f1_knn))
```

```python
cm = confusion_matrix(y_test, y_pred_knn)
print(cm)
```

```python
print(classification_report(y_test, knnmodel.predict(X_test)))
```

# 5. Conclusion
The survey had 20 different features that we can break into two categories. One category are the 10 behavioral ASD screening questions and the other category was 10 individual personal health/traits questions (age/jaundice/ethnicity/parents/country/etc). 

We can see that the 10 behavioral questions are great for helping predict whether or not someone will recieve an ASD diagnosis. It does a much better job predicting in this model than their other personal factors that were acquired during their survey (ethnicity, gender, parent's pdd, etc). The personal factors are important and we can see they do affect a person's likelihood of ASD diagnosis, but they aren't as effective a predictor as the 10 survey questions. 

From the 10 survey questions we can see the overall score is the best predictor for ASD which shows the hallmarks of a competent screening survey. The top three questions that helped predict in and of themselves were 6, 9 and 5. 

Age was the best performing personal factor, but could that possibly be attributed to the older you are the more likely you are to have recieved your diagnosis? I would love to look further into that feature.

In conclusion, the survey does a great job of screening for ASD diagnosis! It could be an excellent screening tool for health professionals to use in guiding them towards next steps. Specifically questions 6, 9, and 5 do an excellent job of helping predict.
